```python
dataset='mnist'

cifar10_classes = {0:'airplane',
               1:'automobile',
               2:'bird',
               3:'cat',
               4:'deer',
               5:'dog',
               6:'frog',
               7:'horse',
               8:'ship',
               9:'truck',}

mnist_classes = {0:'Digit=0',
               1:'Digit=1',
               2:'Digit=2',
               3:'Digit=3',
               4:'Digit=4',
               5:'Digit=5',
               6:'Digit=6',
               7:'Digit=7',
               8:'Digit=8',
               9:'Digit=9',}
    
    
models_mnist = [
    f'xent_mnist',
    f'ntxent_mnist',
    f'tripent_mnist',
    f'trip_mnist',
    f'trip_sup_mnist',
    f'random_init'
]

models_cifar = [
    f'xent_cifar10',
    f'ntxent_cifar10',
    f'tripent_cifar10',
    f'trip_cifar10',
    f'trip_sup_cifar10',
    f'random_init'
]

models_search = [
    'cross-entropy',
    'nt-xent',
    'triplet-entropy',
    'triplet',
    'triplet-supervised',   
    'random', 
    
]

models_nice_name = [
    'Cross-Entropy',
    'NT-XENT',
    'Triplet-Entropy',
    'Triplet-Loss',
    'Triplet-Supervised',
    'Random'
]



cols = ['#3274a1', '#e1812c', '#3a923a', '#c03d3e', '#9372b2', '#000075']
colors = {m:c for m, c in zip(models_nice_name,cols)}
nice_names_mnist = {m:mn for m, mn in zip(models_mnist,models_nice_name)}
nice_names_cifar = {m:mn for m, mn in zip(models_cifar,models_nice_name)}
nice_names_search = {m:mn for m, mn in zip(models_search,models_nice_name)}

def RMQM(data):
    distances = data['average_distance_increase'] + data['average_distance_increase_previous']
    spikes = 1/data['average_spike'] + 1/data['average_spike_previous']
    
    rmqm = distances + spikes
    
    return np.log(1+rmqm)
```
